# **2. Association Rules**

```python
def prepare_crime_data(df):
    """
    Prepare crime data for association rule mining
    """
    # Extract hour from Date
    df['Hour'] = pd.to_datetime(df['Date']).dt.hour

    # Create time periods
    df['time_period'] = pd.cut(
        df['Hour'],
        bins=[0, 6, 12, 18, 24],
        labels=['night', 'morning', 'afternoon', 'evening']
    )

    # Create location categories based on Block field
    df['area'] = df['Block'].str.extract(r'(\d+XX .+)')[0]

    return df

def create_transaction_matrix(df):
    """
    Create binary transaction matrix for association rules
    """
    # Select relevant features for rules
    features_for_rules = [
        'Primary Type',           # Type of crime
        'Location Description',   # Where it occurred
        'time_period',           # Time of day
        'Domestic',              # Domestic incident flag
        'Arrest'                 # Whether arrest was made
    ]

    # Create binary columns
    return pd.get_dummies(df[features_for_rules])
```

```python
def find_rules(transaction_matrix, min_support=0.01, min_confidence=0.3, min_lift=1.1):
    """
    Find association rules with specified thresholds
    """
    # Generate frequent itemsets
    frequent_itemsets = apriori(
        transaction_matrix,
        min_support=min_support,
        use_colnames=True
    )

    # Generate rules
    rules = association_rules(
        frequent_itemsets,
        metric="confidence",
        min_threshold=min_confidence
    )

    # Filter by lift and sort
    rules = rules[rules['lift'] >= min_lift].sort_values('lift', ascending=False)

    return rules
```

```python
def print_rules_summary(rules, top_n=10):
    """
    Print rules in a readable format
    """
    print(f"\nTop {top_n} rules by lift:")
    print("=" * 80)

    for idx, rule in rules.head(top_n).iterrows():
        antecedents = list(rule['antecedents'])
        consequents = list(rule['consequents'])

        print(f"\nRule {idx + 1}:")
        print(f"IF {' AND '.join(antecedents)}")
        print(f"THEN {' AND '.join(consequents)}")
        print(f"Support: {rule['support']:.3f}")
        print(f"Confidence: {rule['confidence']:.3f}")
        print(f"Lift: {rule['lift']:.3f}")
        print("-" * 40)
```

```python
# Main analysis
def analyze_crime_patterns(crime_df):
    """
    Analyze crime patterns using association rules
    """
    print("Preparing data...")
    df_prepared = prepare_crime_data(crime_df)

    print("Creating transaction matrix...")
    transaction_matrix = create_transaction_matrix(df_prepared)

    # Try different support thresholds
    support_thresholds = [0.01, 0.05, 0.1]

    for min_support in support_thresholds:
        print(f"\nAnalyzing patterns with minimum support = {min_support}")
        print("=" * 60)

        rules = find_rules(
            transaction_matrix,
            min_support=min_support,
            min_confidence=0.3,
            min_lift=1.1
        )

        print(f"Found {len(rules)} rules")
        print_rules_summary(rules, top_n=5)

        # Count feature types in rules
        antecedent_features = []
        consequent_features = []

        for items in rules['antecedents']:
            antecedent_features.extend([item.split('_')[0] for item in items])

        for items in rules['consequents']:
            consequent_features.extend([item.split('_')[0] for item in items])

        print("\nMost common features in rule conditions:")
        for feature, count in pd.Series(antecedent_features).value_counts().head().items():
            print(f"{feature}: {count} rules")

        print("\nMost common features in rule predictions:")
        for feature, count in pd.Series(consequent_features).value_counts().head().items():
            print(f"{feature}: {count} rules")

    return rules

# Run the analysis
rules = analyze_crime_patterns(crime_df)
```

# **3. Additional Ward-based Rules**

```python
def analyze_ward_patterns(crime_df):
    """
    Simple ward-focused association rules analysis
    """
    # Create a copy to avoid modifying original
    df = crime_df.copy()

    # Convert Ward to string category to handle as dummy variable
    df['Ward'] = 'Ward_' + df['Ward'].astype(str)

    # Create dummy variables
    ward_dummies = pd.get_dummies(df['Ward'])
    type_dummies = pd.get_dummies(df['Primary Type'])
    location_dummies = pd.get_dummies(df['Location Description'])

    # Combine all dummy variables
    df_encoded = pd.concat([ward_dummies, type_dummies, location_dummies, df['Domestic']], axis=1)

    # Find frequent itemsets
    print("Mining frequent itemsets...")
    frequent_itemsets = apriori(
        df_encoded,
        min_support=0.01,  # 1% minimum support
        use_colnames=True
    )

    # Generate rules
    print("Generating rules...")
    rules = association_rules(
        frequent_itemsets,
        metric="confidence",
        min_threshold=0.3  # 30% minimum confidence
    )

    # Filter for ward-related rules
    ward_rules = rules[
        rules['antecedents'].apply(lambda x: any('Ward_' in str(i) for i in x)) |
        rules['consequents'].apply(lambda x: any('Ward_' in str(i) for i in x))
    ]

    # Sort by lift and filter for meaningful rules
    ward_rules = ward_rules[ward_rules['lift'] > 1.1].sort_values('lift', ascending=False)

    print(f"\nFound {len(ward_rules)} ward-related rules")
    print("\nTop 10 strongest ward-related patterns:")
    print("=" * 80)

    for idx, rule in ward_rules.head(10).iterrows():
        print(f"\nRule {idx + 1}:")
        print(f"IF {list(rule['antecedents'])}")
        print(f"THEN {list(rule['consequents'])}")
        print(f"Support: {rule['support']:.3f}")
        print(f"Confidence: {rule['confidence']:.3f}")
        print(f"Lift: {rule['lift']:.3f}")

    # Additional ward insights
    print("\nWard Crime Distribution (Top 10):")
    ward_crime_dist = df.groupby('Ward')['Primary Type'].value_counts()
    print(ward_crime_dist.head(10))

    print("\nWard Location Distribution (Top 10):")
    ward_loc_dist = df.groupby('Ward')['Location Description'].value_counts()
    print(ward_loc_dist.head(10))

    return ward_rules

# Run the analysis
ward_rules = analyze_ward_patterns(crime_df)
```
